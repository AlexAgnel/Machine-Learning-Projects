```python
# Importing libraries
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_csv("dataset.csv")
dataset.head()
```

```python
# Summaraize the dataset
dataset.shape
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Segregate dataset into x and y
x = dataset.iloc[:,:-1].values
x
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
# Splitting dataset into train and test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
# Predicting and Validating
y_pred = model.predict(x_test)

from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)

r2_score = r2_score(y_test,y_pred)
print("R_Squared Score:",r2_score)
```

```python
# Visualization of Decision Tree Regressor
import matplotlib.pyplot as plt

x_val = np.arange(min(x_train),max(x_train),0.01)
x_val = x_val.reshape((len(x_val),1))
plt.scatter(x_train, y_train, color='green')
plt.plot(x_val, model.predict(x_val), color='red')
plt.title("Height prediction using DecisionTree")
plt.xlabel("Age")
plt.ylabel("Height")
plt.figure()
plt.show()
```
